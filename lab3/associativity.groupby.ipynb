```python
import pandas as pd
```

```python
infile = 'associativity.trace2.csv'
outfile = 'associativity.trace2.groupby.xlsx'
```

```python
df = pd.read_csv(infile)
df
```

```python
df = df[['Associativity', 'CacheSize', 'Miss Rate']]
df
```

```python
grouped = df.groupby(['CacheSize', 'Associativity']).mean()
grouped
```

```python
Associativiys = df['Associativity'].unique()
Associativiys
```

```python
cache_sizes = df['CacheSize'].unique()
cache_sizes
```

```python
new_df = pd.DataFrame(index=cache_sizes, columns=Associativiys)
new_df
```

```python
for cache_size in cache_sizes:
    for associativiy in Associativiys:
        new_df.loc[cache_size,associativiy] = grouped.loc[cache_size,associativiy][0]
```

```python
new_df
```

```python
new_df.to_excel(outfile)
```
